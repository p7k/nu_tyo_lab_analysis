```python
def cmyk_to_rgb(c, m, y, k):
    r = 255 * (1 - c) * (1 - k)
    g = 255 * (1 - m) * (1 - k)
    b = 255 * (1 - y) * (1 - k)
    return [int(chan) for chan in [r, g, b]]
```

```python
def format_for_tex(row: tuple, color_sys='RGB') -> str:
    sre = re.search(r'RGB|CMYK', row['name'], re.I)
    if sre:
        color_sys = sre.group()
        name = re.sub(f'{color_sys}', '', name, re.I)
    name = re.sub(r'[\s-]+', '', row['name'], re.I)
    if color_sys:
    values = [(round(v, 4)) for v in row['data']['values']]
    return f'\\definecolor{{{name}}}{{{color_sys}}}{{{','.join(map(str, values))}}}]'
```

```python
df_swatch_rgb_secondary.apply(format_for_tex, axis=1)
```
